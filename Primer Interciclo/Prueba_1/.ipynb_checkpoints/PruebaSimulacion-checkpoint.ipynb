#### Generar gráfico de pie por fabricante de la vacuna.


```python
datos_fabri=pd.read_csv('https://raw.githubusercontent.com/andrab/ecuacovid/master/datos_crudos/vacunas/fabricantes.csv', \
header=None)
datos_fabri.columns = ['vaccine', 'total', 'arrived_at']
a = np.array(list(set(datos_fabri['vaccine'][1:])))
print(a)

oxford = 0
sino = 0
pfizer = 0
for i, j in zip(datos_fabri['vaccine'][1:], datos_fabri['total'][1:]):
    if i == a[0]:
  
        sino = sino + + int(j)
    elif i == a[1]:
        pfizer = pfizer + + int(j)
    elif i == a[2]:
        oxford = oxford + int(j)

b = np.array([sino, pfizer, oxford])
print(b)
plt.pie(b, labels=a, autopct="%0.1f %%")
plt.gcf().set_size_inches(50, 10)
plt.show()
```

### ***Comparacion de Ecuador y Colombia***

```python
datos = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv'
df_vacum = pd.read_csv(datos, header = None).fillna(0)
df_vacum.columns = ['location', 'iso_code', 'date', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', \
                    'daily_vaccinations_raw', 'daily_vaccinations', 'total_vaccinations_per_hundred', \
                    'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'daily_vaccinations_per_million']
lista_fechas = []
lista_tot = []
lista_fechasbr = []
lista_totbr = []
for i, j, k in zip(df_vacum['location'], df_vacum['date'], df_vacum['people_fully_vaccinated']):
#   print(i)
    if i == 'Ecuador':
        lista_fechas.append(j)
        lista_tot.append(k)
    elif i == 'Colombia':
        lista_fechasbr.append(j)
        lista_totbr.append(k)
dat = np.array(lista_fechas) 
tot = np.array(lista_tot, dtype='float')
x = np.arange(1, len(tot) + 1, 1) 
dat_br = np.array(lista_fechasbr) 
tot_br = np.array(lista_totbr, dtype='float')
xbr = np.arange(1, len(tot_br) + 1, 1) 
from sklearn import linear_model 
print("-----------------Ecuador--------------") 
regr = linear_model.LinearRegression()  
regr.fit(np.array(x).reshape(-1, 1), tot)
print('Coefficients: \n', regr.coef_)
print('Independent term: \n', regr.intercept_)
print("----------------Colombia---------------") 
regrbr = linear_model.LinearRegression()
regrbr.fit(np.array(xbr).reshape(-1, 1), tot_br)
print('Coefficients: \n', regrbr.coef_)
print('Independent term: \n', regrbr.intercept_)
#SE GRAFICA
fig, (ax1, ax2) = plt.subplots(1, 2, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0}, figsize=(15,10))
fig.suptitle('Comparación de Vacunación entre los paises Ecuador-Colombia',color='green',size='25')
ax1.scatter(x, tot, color='violet')
ax1.set_title('Ecuador')
x_real = np.array(range(50, 100))
ax2.scatter(xbr, tot_br, color='red')
ax2.set_title('Colombia')
```
