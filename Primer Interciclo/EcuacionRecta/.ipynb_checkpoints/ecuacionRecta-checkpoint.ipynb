## Encontrar la ecuacion de la recta dado dos puntos

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

def generar_datos():
    x = np.random.random(20)*10
    y = np.random.random(20)*10
    return x.reshape((20,1)), y.reshape((20,1))
x,y = generar_datos()

plt.plot(x, y,'x')
plt.grid(True)
plt.show()
```

```python
modelo = linear_model.LinearRegression()
modelo.fit(x, y)
y_pred = modelo.predict(x)
print(modelo.coef_[0])
b = -modelo.coef_[0][0]*x[0]+y_pred[0]
plt.scatter(x, y)
plt.xlim([-5, 20])
plt.ylim([-5, 20])
plt.plot(x, y_pred, color='red')
plt.grid(True)
x_real = np.array([0, 100])
plt.show()
if (b < 0):
    ecua ='y = {} x {}'
else:
    ecua ='y = {} x + {}'
print('La ecuacion de la recta es: ')
print(ecua.format(modelo.coef_[0][0],b[0]))
```
